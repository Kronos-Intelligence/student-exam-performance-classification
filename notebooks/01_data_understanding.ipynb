# Missing Values Assessment

Missing values can introduce bias, reduce model performance, and impact the reliability of statistical analyses. As part of the initial data quality assessment, each variable is evaluated to determine whether any observations contain missing information.

This assessment supports data quality management and helps determine whether imputation or other preprocessing techniques will be required.

```python
# Count missing values
missing_values = df.isnull().sum()

missing_values
```

```python
# Create a missing value summary
missing_summary = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percent Missing": (df.isnull().sum() / len(df)) * 100
})

missing_summary
```

## Interpretation

No missing values were identified in the dataset. Every variable contains complete observations across all 10,000 student records.

From a data quality perspective, this indicates a high level of completeness and eliminates the need for missing value imputation during preprocessing. While this simplifies model development, additional quality checks—including duplicate detection, outlier identification, and validation of categorical values—will still be performed in later stages of the project.

# Duplicate Records Assessment

Duplicate records can negatively impact statistical analyses and machine learning models by overrepresenting certain observations. As part of the initial data quality assessment, the dataset is evaluated to determine whether duplicate student records are present.

Identifying duplicate records helps ensure that each observation represents a unique student and contributes appropriately to the analysis.

```python
# Count duplicate records
duplicate_records = df.duplicated().sum()

print(f"Duplicate Records: {duplicate_records}")
```

```python
# Display duplicate records if any exist
if duplicate_records > 0:
    display(df[df.duplicated()])
else:
    print("No duplicate records were found.")
```

## Interpretation

The duplicate record assessment identified no duplicate observations within the dataset. Each record appears to represent a unique student, indicating a high level of data uniqueness.

From a data governance perspective, uniqueness is one of the core dimensions of data quality. Maintaining unique records helps ensure accurate reporting, reliable predictive modeling, and trustworthy analytical outcomes.
